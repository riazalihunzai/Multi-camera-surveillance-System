```python
import cv2
import numpy as np
import time
import argparse

```

```python
net = cv2.dnn.readNet("yolov3-tiny.weights","yolov3-tiny.cfg")
classesFile = "coco.names";
classes = None
with open(classesFile, 'rt') as f:
    classes = f.read().rstrip('\n').split('\n')
```

```python
layer_names = net.getLayerNames()
outputlayers = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()]
```

```python
colors= np.random.uniform(0,255,size=(len(classes),3))
```

```python
def box(blob,frame,font):
                height,width,channels = frame.shape
                net.setInput(blob)
                outs = net.forward(outputlayers)
                class_ids=[]
                confidences=[]
                boxes=[]
                for out in outs:
                    for detection in out:
                        scores = detection[5:]
                        class_id = np.argmax(scores)
                        confidence = scores[class_id]
                        if confidence > 0.3:

                            center_x= int(detection[0]*width)
                            center_y= int(detection[1]*height)
                            w = int(detection[2]*width)
                            h = int(detection[3]*height)
                            x=int(center_x - w/2)
                            y=int(center_y - h/2)


                            boxes.append([x,y,w,h]) 
                            confidences.append(float(confidence)) 
                            class_ids.append(class_id) 
                
                indexes = cv2.dnn.NMSBoxes(boxes,confidences,0.3,0.4)
                for i in range(len(boxes)):
                    if i in indexes and class_ids[i]==0:
                        print ("Person is detected")
                        x,y,w,h = boxes[i]
                        label = str(classes[class_ids[i]])
                        confidence= confidences[i]
                        color = colors[class_ids[i]]
                        cv2.rectangle(frame,(x,y),(x+w,y+h),color,2)
                        cv2.putText(frame,label+" "+str(round(confidence,2)),(x,y+30),font,1,(255,255,255),2)
            
                return frame
```

```python
def offline_mode(filename1,filename2,filename3):
        #reading files
        cap_1=cv2.VideoCapture(filename1)
        cap_2=cv2.VideoCapture(filename2)
        cap_3=cv2.VideoCapture(filename3) 
        font = cv2.FONT_HERSHEY_PLAIN
        #Writing to files
        frame_width_1 = int(cap_1.get(3)) 
        frame_height_1 = int(cap_1.get(4))
        size_1=(frame_width_1,frame_height_1)

        frame_width_2 = int(cap_2.get(3)) 
        frame_height_2 = int(cap_2.get(4))
        size_2=(frame_width_2,frame_height_2)

        frame_width_3 = int(cap_3.get(3)) 
        frame_height_3 = int(cap_3.get(4))
        size_3=(frame_width_3,frame_height_3)
        fourcc = cv2.VideoWriter_fourcc(*'XVID')
        wrt_1= cv2.VideoWriter('BoxedVideo/Offline/IpWeb_1.avi',fourcc, 20.0, size_1)
        wrt_2= cv2.VideoWriter('BoxedVideo/Offline/IpWeb_2.avi',fourcc, 20.0, size_2)
        wrt_3= cv2.VideoWriter('BoxedVideo/Offline/IpWeb_3.avi',fourcc, 20.0, size_3)    
        while True:
            
            ret_1,frame_1=cap_1.read()  
            ret_2,frame_2=cap_2.read()
            ret_3,frame_3=cap_3.read()
            if   ret_1==True and ret_2==True and ret_3==True:
                
                
                
                blob1 = cv2.dnn.blobFromImage(frame_1,0.00392,(320,320),(0,0,0),True,crop=False)   
                blob2 = cv2.dnn.blobFromImage(frame_2,0.00392,(320,320),(0,0,0),True,crop=False)     
                blob3 = cv2.dnn.blobFromImage(frame_3,0.00392,(320,320),(0,0,0),True,crop=False)         
                frame_1=box(blob1,frame_1,font)
                frame_2=box(blob2,frame_2,font)
                frame_3=box(blob3,frame_3,font)
                
                    
                cv2.imshow("webcam1",frame_1)
                cv2.imshow("webcam2",frame_2)
                cv2.imshow("webcam3",frame_3)
                wrt_1.write(frame_1)
                wrt_2.write(frame_2)
                wrt_3.write(frame_3)
                key = cv2.waitKey(1) 

                if key == 27:
                    break;
            else :
                break
        cap_1.release()
        cap_2.release()
        cap_3.release()
        wrt_1.release()
        wrt_2.release()
        wrt_3.release()
        cv2.destroyAllWindows()
```

```python
offline_mode("IpWeb_1.avi","IpWeb_2.avi","IpWeb_3.avi")
```
